```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv("C:\\Users\\ankit\\Downloads\\branch-wise-oper-perf-2021-22.csv")

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['SANCTIONS Rs in crore'], df['DISBURSEMENT Rs in crore'])
plt.xlabel('SANCTIONS (Rs in c"rore)')
plt.ylabel('DISBURSEMENT (Rs in crore)')
plt.title('Relationship between SANCTIONS and DISBURSEMENT')
plt.show()

# Bar chart
# We'll use the first 10 offices for better visibility
top_10_offices = df.nlargest(10, 'RECOVERY Rs  in crore')

plt.figure(figsize=(12, 6))
plt.bar(top_10_offices['OFFICES'], top_10_offices['RECOVERY Rs  in crore'])
plt.xlabel('Offices')
plt.ylabel('RECOVERY (Rs in crore)')
plt.title('Top 10 Offices by Recovery Amount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Calculate correlation coefficient
correlation = df['SANCTIONS Rs in crore'].corr(df['DISBURSEMENT Rs in crore'])
print(f"Correlation coefficient between SANCTIONS and DISBURSEMENT: {correlation:.4f}")

# Calculate mean recovery
mean_recovery = df['RECOVERY Rs  in crore'].mean()
print(f"Mean RECOVERY across all offices: {mean_recovery:.2f} Rs in crore")
```
