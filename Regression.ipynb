```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Read the CSV file
df = pd.read_csv("C:\\Users\\ankit\\Downloads\\branch-wise-oper-perf-2021-22.csv")

# Prepare the data
X = df['SANCTIONS Rs in crore'].values.reshape(-1, 1)
y = df['DISBURSEMENT Rs in crore'].values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Evaluate the model
r_squared = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

# Print results
print("Linear Regression Results:")
print(f"Intercept (β₀): {model.intercept_:.4f}")
print(f"Coefficient (β₁): {model.coef_[0]:.4f}")
print(f"R-squared: {r_squared:.4f}")
print(f"Mean Squared Error: {mse:.4f}")

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('SANCTIONS Rs in crore')
plt.ylabel('DISBURSEMENT Rs in crore')
plt.title('Linear Regression: DISBURSEMENT vs SANCTIONS')
plt.legend()
plt.show()

# Plot residuals
residuals = y - y_pred
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals)
plt.xlabel('Predicted DISBURSEMENT')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()

# Interpret coefficients
print("\nInterpretation of coefficients:")
print(f"For every 1 crore Rs increase in SANCTIONS, DISBURSEMENT increases by {model.coef_[0]:.4f} crore Rs on average.")
print(f"When SANCTIONS is 0, the expected DISBURSEMENT is {model.intercept_:.4f} crore Rs.")

# Evaluate assumptions
print("\nEvaluation of assumptions:")
print("1. Linearity: Check if the scatter plot shows a linear relationship.")
print("2. Independence: Assume observations are independent based on data collection.")
print("3. Homoscedasticity: Check if residual plot shows consistent spread.")
print("4. Normality: Check if residuals are approximately normally distributed.")
print("5. No multicollinearity: Not applicable for simple linear regression with one predictor.")
```
